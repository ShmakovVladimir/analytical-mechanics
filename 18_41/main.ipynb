```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as smp
```

```python
a1, a2, a3, m, g, p, l, t, lamb, c = smp.symbols(r"a_{1} a_{2} a_{3} m g p l t \lambda c")

x1, x2, x3 = smp.symbols("x_{1} x_{2} x_{3}", cls = smp.Function)
x1 = x1(t)
x2 = x2(t)
x3 = x3(t)

x1_d, x2_d, x3_d = smp.diff(x1, t), smp.diff(x2, t), smp.diff(x3, t)

x1_dd, x2_dd, x3_dd = smp.diff(x1_d, t), smp.diff(x2_d, t), smp.diff(x3_d, t)

q = smp.Matrix([x1, x2, x3])
q_d = smp.Matrix([x1_d, x2_d, x3_d])
```

```python
T = m * (x1_d**2 + x2_d**2 + x3_d**2) / 2
P = c * ((x1 - x2)**2) / 2 + c * ((x3 - x2)**2) / 2 - m*g*(smp.sqrt(l**2 - x1**2) + smp.sqrt(l**2 - x2**2) + smp.sqrt(l**2 - x3**2))
P
```

```python
A = smp.Matrix([[smp.diff(smp.diff(T, q_d[j]), q_d[i]) for i in range(len(q))] for j in range(len(q))])         
A               
```

```python
C = smp.Matrix([[smp.diff(smp.diff(P, q[j]), q[i]) for i in range(len(q))] for j in range(len(q))])
C
```

```python
charEq = C - lamb * A
charEq
```

```python
eq = charEq[0, 0] * charEq[1, 1] * charEq[2, 2] - (c**2) * charEq[0, 0] - (c**2) * (charEq[2, 2])
sols = smp.solve([eq], [lamb])
```
